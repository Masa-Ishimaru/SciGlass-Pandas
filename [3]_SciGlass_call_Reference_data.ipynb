## Load a file with columns of Kod, GlasNo, composition, and properties

```python
# This is just an example dataframe

SciGlass_comp_prop = pd.read_csv('data_SciGlass/SciGlass_comp_prop.csv')
print(SciGlass_comp_prop.shape)
SciGlass_comp_prop.head()
```

## Extract Kod, Refer_Id data from the "ARHIV" file

```python
ARHIV = pd.read_csv('data_SciGlass/ARHIV.csv', encoding = 'latin1')
print(ARHIV.shape)
ARHIV.head()
```

```python
# Use only Kod and Refer_ID

ARHIV_select = ARHIV[['KOD', 'Refer_Id']]
ARHIV_select = ARHIV_select.rename(columns={'KOD': 'Kod', 'Refer_Id': 'Refer_ID'})
print(ARHIV_select.shape)
ARHIV_select.head()
```

## Extract Refer_Id, YEAR, JOURNAL, Authors, Title, Discr data from the "Reference" file

```python
Reference = pd.read_csv('data_SciGlass/Reference.csv', encoding = 'latin1')
print(Reference.shape)
Reference.head()
```

```python
Reference_select = Reference[['Refer_ID', 'YEAR', 'JOURNAL', 'Authors', 'Title', 'Descr']]
print(Reference_select.shape)
Reference_select.head()
```

## Combine Kod with the source data from "Reference" file on Refer_ID

```python
SciGlass_source = pd.merge(ARHIV_select, Reference_select, on = 'Refer_ID', how = 'inner')
print(SciGlass_source.shape)
SciGlass_source.head()
```

## Combine "SciGlass_comp_prop" file with the source data on Kod

```python
SciGlass_select_source = pd.merge(SciGlass_comp_prop, SciGlass_source, on = 'Kod', how = 'inner')
print(SciGlass_select_source.shape)
SciGlass_select_source.head(20)
```

## Sort by YEAR if necessary

```python
SciGlass_select_source_s = SciGlass_select_source.sort_values('YEAR')
print(SciGlass_select_source_s.shape)
SciGlass_select_source_s.head(20)
```
